```python
mapping = {"Iris-versicolor":5,"Iris-virginica":4,"Iris-setosa":3}
df["Species"] = df["Species"].map(mapping)
```

```python
df.head()
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
```

```python
df
```

```python
df["Species"] = le.inverse_transform(df['Species'])   # revert back to the original values
```
