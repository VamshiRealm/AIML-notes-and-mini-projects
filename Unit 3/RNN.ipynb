```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/vamsh/Downloads/smsspamcollection.tsv", delimiter= '\t')
```

```python
df.head()
```

```python
x = df.message
y = df.label
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.2, random_state=1, shuffle=True)
```

# Text preprocessing

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
max_words = 5000  #vocabluary size
max_len = 100  #sequence legth

token = Tokenizer(num_words = max_words)  #identifying the tokens
token.fit_on_texts(xtrain)

xtrain_seq = token.texts_to_sequences(xtrain) #converting tokens into sequence
xtest_seq = token.texts_to_sequences(xtest)

xtrain_pad = pad_sequences(xtrain_seq, maxlen = max_len)  #converting sequence into same length
xtest_pad = pad_sequences(xtest_seq, maxlen = max_len)
```

# RNN

```python
from keras.models import Sequential
from keras.layers import Embedding, Dense, SimpleRNN
```

```python
model = Sequential()
model.add(Embedding(input_dim=max_words, output_dim = 64))
model.add(SimpleRNN(64))
model.add(Dense(36, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(xtrain_pad, ytrain, epochs=20)
```

```python
model.evaluate(xtest_pad, ytest)
```

```python
ypred = model.predict(xtest_pad)
```

```python
l1 = []
for i in ypred:
    if i > 0.5:
        l1.append(1)
    else:
        l1.append(0)
```

# LSTM

```python
from keras.models import Sequential
from keras.layers import Embedding, Dense, LSTM
```

```python
model1 = Sequential()
model1.add(Embedding(input_dim=max_words, output_dim = 64))
model1.add(LSTM(64))
model1.add(Dense(36, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model1.fit(xtrain_pad, ytrain, epochs=20)
```

```python
model1.evaluate(xtest_pad, ytest)
```

```python
ypred = model1.predict(xtest_pad)
```

```python
l2 = []
for i in ypred:
    if i > 0.5:
        l2.append(1)
    else:
        l2.append(0)
```
