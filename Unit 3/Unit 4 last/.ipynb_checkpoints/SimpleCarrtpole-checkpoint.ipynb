```python
import random as r
import gym
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Invoking the builtIn environment
```

```python
env = gym.make('CartPole-v1', render_mode="human")
```

```python
# Learning based on states
```

```python
episodes = 50
for episode in range(1, episodes):
    state = env.reset()
    done = False
    score = 0
    while not done:
        action = r.choice([0,1])
        n_state, reward, done, truncated, info = env.step(action)  #n_state, reward, done are compulsory 

        score += reward
        env.render()
    print("Episode: ",episode," Score: ",score)
env.close()

    
    
```
