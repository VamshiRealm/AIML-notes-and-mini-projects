```python
import pandas as pd
df = pd.read_csv("C:/Users/vamsh/Downloads/income.csv")
```

```python
from sklearn.mixture import GaussianMixture
gmm=GaussianMixture(n_components = 2)
gmm.fit(df)
```

```python
labels = gmm.predict(df)
```

```python
df['labels'] = labels
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['INCOME'],df['SPEND'], c=df['labels'])
plt.show()
```
