# Perceptron Model

```python
import pandas as pd
df = pd.DataFrame({"x1":[0,0,1,1],"x2":[0,1,0,1],"y":[0,1,1,0]})  #AND GATE
df.head()
```

```python
from sklearn.linear_model import Perceptron
```

```python
x = df[["x1","x2"]]
y = df.y
model = Perceptron()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.neural_network import MLPClassifier
nn= MLPClassifier(hidden_layer_sizes=(50,10), activation="relu")  # Sigmoid is called logistic
nn.fit(x,y)
```

```python
nn.score(x,y)
```

```python
ypred = nn.predict(x)
ypred
```
