```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
model= Sequential()
#First Convolution
model.add(Conv2D(32,(3,3), input_shape=(64,64,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))  #optional  (used for avoiding overfitting problem)

#Second Convolution
model.add(Conv2D(16,(3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25)) 

#Flattening (Input layer for MLP)
model.add(Flatten())

#Fully connected layer
model.add(Dense(32,activation="relu"))
model.add(Dense(8,activation="relu"))
model.add(Dense(1,activation="sigmoid"))


```

```python
model.compile(loss="binary_crossentropy", metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen = ImageDataGenerator(rescale=1.0/255, shear_range= 0.2, zoom_range=0.2, horizontal_flip=True) 
test_gen = ImageDataGenerator(rescale=1.0/255) 
#rescale : sclaing the size of pixels, #shear: change the size
#shear_range, zoom_range, horizontal_flip: all are the parameters of data augmentation
```

```python
train_set = train_gen.flow_from_directory("C:/Users/vamsh/Downloads/image/train", target_size=(64,64), class_mode='binary', batch_size=32)
```

```python
model.fit(train_set, epochs = 25)
```
