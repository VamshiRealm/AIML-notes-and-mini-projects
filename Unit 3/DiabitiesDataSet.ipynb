```python
import pandas as pd
df = pd.read_csv("C:/Users/vamsh/Downloads/diabetes_indian.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
x = df.iloc[:,0:8]
y = df.Outcome
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
x = mm.fit_transform(x)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain , ytest = train_test_split(x, y , test_size = 0.2, shuffle = True, random_state = 1)
```

# MLP Classifier

```python
from sklearn.neural_network import MLPClassifier
nn_mlp = MLPClassifier(hidden_layer_sizes=(50,50), activation="relu")
nn_mlp.fit(xtrain,ytrain)
```

```python
print("Training Accuracy: ", nn_mlp.score(xtrain, ytrain))
print("Testing Accuracy: ", nn_mlp.score(xtest,ytest))
```

```python
ypred = nn_mlp.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,ypred)
```

```python
###Tensorflow and keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(16,input_dim=8, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(1, activation="sigmoid"))   #sigmoid is sigmoid in keras
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])  #loss parameter is compulsory
```

```python
model.fit(xtrain, ytrain, epochs = 25)
```

```python
ypred = model.predict(xtest)
```

```python
predicted  = []
for i in ypred:
    if i > 0.5:
        predicted.append(1)
    else:
        predicted.append(0)
print(predicted)
```

```python
model.evaluate(xtest, ytest)
```
