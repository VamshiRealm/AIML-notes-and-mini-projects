```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/vamsh/Downloads/BMI.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender']= le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
l1 = ["Height","Weight"]
for i  in l1:
    df[i] = mm.fit_transform(df[[i]])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df[["Gender","Height","Weight"]]
y = df["Index"]
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.8, random_state = 1, shuffle=True)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
gnb = GaussianNB()
dt = DecisionTreeClassifier()
```

```python
lr.fit(xtrain, ytrain)
gnb.fit(xtrain, ytrain)
dt.fit(xtrain, ytrain)
```

# Evaluate all the accuracy

```python
print("Logistic Regression: ",lr.score(xtest, ytest))
print("GaussianNB: ",gnb.score(xtest, ytest))
print("DecisionTree: ",dt.score(xtest, ytest))
```

# OverFitting and UnderFitting

```python
print("Training Accuracy = ", dt.score(xtrain, ytrain))
print("Testing Accuracy = ", dt.score(xtest, ytest))
```

# Checking all other metrics

```python
ypred = dt.predict(xtest)
from sklearn.metrics import confusion_matrix, precision_score, recall_score
print(confusion_matrix(ytest,ypred))
```

```python
df.Index.value_counts()
```

```python
print(precision_score(ytest,ypred,average="weighted" ))
```

```python
print(recall_score(ytest,ypred, average="weighted"))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

```python
from sklearn.metrics import roc_auc_score
yprob = dt.predict_proba(xtest)
yprob
```

```python
print(roc_auc_score(ytest,yprob, multi_class="ovr"))
```
