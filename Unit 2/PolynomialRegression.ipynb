```python
import pandas as pd
df = pd.read_csv("C:/Users/vamsh/Downloads/Housing.csv")
```

```python
# feature Scaling
l2 = ['price','lotsize']
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
for i in l2:
    df[i] = mm.fit_transform(df[[i]])
```

```python
# Label Encoder
l1 = ['driveway','recroom', 'fullbase', 'gashw', 'airco', 'prefarea']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in l1:
    df[i] = le.fit_transform(df[i])
    
df.head()
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

# Polynomial Regression

```python
x = df[['lotsize']]
y = df['price']
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 2)
xpoly = pf.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(xpoly, y, test_size = 0.2, random_state = 1, shuffle = True)
```

```python
# ypred = pr.predict(xtest)
# from sklearn.metrics import r2_score
# r2_score(ytest, ypred)
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(xtrain, ytrain)
print("m = ",slr.coef_)
print("b = ",slr.intercept_)
```

```python
ypred = slr.predict(xtest)
from sklearn.metrics import r2_score
r2_score(ytest, ypred)
```

# Solution for Overfitting

# <h2> Shrinkage Methods </h2>

```python
print("Training Accuracy = ", slr.score(xtrain, ytrain))   # .score acts same as r2_score but it is more easy to use
print("Testing Accuracy = ", slr.score(xtest, ytest))
```

```python
from sklearn.linear_model import Lasso  # Lasso brings the value of (m) intercept to zero
l1 = Lasso()
l1.fit(xtrain, ytrain)
print("m = ",l1.coef_, " and b = ", l1.intercept_)
```

```python
from sklearn.linear_model import Ridge  # Ridge brings the value of (m) intercept near to each other
l2 = Ridge()
l2.fit(xtrain, ytrain)
print("m = ",l2.coef_, " and b = ", l2.intercept_)
```

```python
from sklearn.linear_model import ElasticNet  # Applies both l1 and l2 regression
l3 = ElasticNet()
l3.fit(xtrain, ytrain)
print("m = ",l3.coef_, " and b = ", l3.intercept_)
```

```python
print("Lasso Accuracy = ", l1.score(xtest, ytest))
print("Ridge Accuracy = ", l2.score(xtest, ytest))
print("ElasticNet Accuracy = ", l3.score(xtest, ytest))
```
