```python
import pandas as pd
df = pd.read_csv("C:/Users/vamsh/Downloads/Housing.csv")
```

```python
df.head()
```

```python
del df['Unnamed: 0']
df.isna().sum()
```

```python
# feature Scaling
l2 = ['price','lotsize']
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
for i in l2:
    df[i] = mm.fit_transform(df[[i]])
```

```python
# Label Encoder
l1 = ['driveway','recroom', 'fullbase', 'gashw', 'airco', 'prefarea']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in l1:
    df[i] = le.fit_transform(df[i])
    
df.head()
```

```python
x = df[['lotsize']]
y = df[['price']]
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.75, random_state = 1, shuffle = True)
```

# Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(xtrain, ytrain)
```

```python
print("m = ",slr.coef_)
```

```python
print("b = ",slr.intercept_)
```

# Evaluation Phase

```python
ypred = slr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest, ypred)
```

# Multiple Linear Regression

```python
x = df[['lotsize', 'bedrooms']]
y = df[['price']]
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.75, random_state = 1, shuffle = True)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(xtrain, ytrain)
print("m = ",mlr.coef_)
print("b = ",mlr.intercept_)
```
