```python
import pandas as pd
df = pd.read_csv("C:/Users/vamsh/Downloads/Housing.csv")
```

```python
# feature Scaling
l2 = ['price','lotsize']
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
for i in l2:
    df[i] = mm.fit_transform(df[[i]])
```

```python
# Label Encoder
l1 = ['driveway','recroom', 'fullbase', 'gashw', 'airco', 'prefarea']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in l1:
    df[i] = le.fit_transform(df[i])
    
df.head()
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

# Polynomial Regression

```python
x = df[['lotsize']]
y = df['price']
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 2)
xpoly = pf.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(xpoly, y, test_size = 0.2, random_state = 1, shuffle = True)
```

```python
from sklearn.linear_model import LinearRegression
pr = LinearRegression()
pr.fit(xtrain, ytrain)
print("m = ",pr.coef_)
print("b = ",pr.intercept_)
```

```python
ypred = pr.predict(xtest)
from sklearn.metrics import r2_score
r2_score(ytest, ypred)
```
